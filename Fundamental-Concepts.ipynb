BIAS :

    1.Overfitting
    2.When high , the model is simple
    3.Types:
        Algorithmic Bias: This type of bias arises from the algorithms and techniques used in the machine learning process. It can occur due to various reasons, such as the choice of features, the model's architecture, or the optimization strategy employed during training. For example, if a model is designed in a way that favors certain patterns in the data or ignores other relevant information, it may exhibit algorithmic bias.

        Data Bias: Data bias refers to the presence of skewed or unrepresentative data in the training dataset. If the training data is not diverse and does not fully capture the real-world distribution of the problem, the model may not generalize well to unseen examples, leading to biased predictions. Data bias can arise from many sources, including imbalanced class distribution, data collection methods, or sampling biases.

VARIANCE : 

    1.Underfitting
    2.When high , the model is complex
    3.The variation in distribution of data in the given dataset is called variance .



OVERFITTING AND UNDERFITTING :

OVERFITTING:

    1.The model memorizes the data
    2.The model can't generalize the testing data well enough 
    3.The model overlearns the underlying data concepts

UNDERFITTING:

    1.The model cannot understand the data concept underlying in it
    2.It cannot perform well on the training data itself
